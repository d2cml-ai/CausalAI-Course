8.2. The flexible model will have the same columns, and will also include polynomials for experience (exp2 exp3 exp4), as well as the interactions between all experience variables and other variables except for sex. 

Hint: you can use a for loop to multiply the desired variables and create the interactions. Some packages might also have ways to easily specify the model variables with strings like R does by default

$
\text{sex} + \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} +
\text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + 
\text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} +
\left( \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} \right) \times 
\left( \text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + 
\text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} \right)
$

```python
#variables
predictors_2 = sin_salario[['sex', 'exp1', 'exp2', 'exp3', 'exp4','hsg', 'scl', 'clg', 'ad', 'so', 'we', 'ne', 'occ2', 'ind2']]
#obtenemos dummies
predictors_2 = pd.get_dummies(predictors_2, columns=['occ2', 'ind2'], drop_first=True, dtype=int)

# suma se exp1 + ... exp4
predictors_2['exp'] = predictors_2[['exp1', 'exp2', 'exp3', 'exp4']].sum(axis=1)
# suma de las demas variables expeto las dummies y sex
predictors_2['suma'] = predictors_2[['hsg', 'scl', 'clg', 'ad', 'so', 'we', 'ne']].sum(axis=1)
# suma de dummies
dummy_columns = [col for col in predictors_2.columns if col.startswith('occ2_') or col.startswith('ind2_')]
predictors_2['dummy_sum'] = predictors_2[dummy_columns].sum(axis=1)

# suma de la sum de dummies de occ2 y ind2, con la suma de las demás variables
predictors_2['sum_2'] = predictors_2[['suma', 'dummy_sum']].sum(axis=1)

# variable de interacción
predictors_2['interaction'] = predictors_2['exp'] * predictors_2['sum_2']

#elimino variables que se usaron para obtener la variable de interacción
predictors_2 = predictors_2.drop(columns=['exp','suma', 'dummy_sum','sum_2'])

display(predictors_2.columns)
display(predictors_2.shape)
```

8.3. The extra-flexible model will include all two-way interactions between variables, except for sex. 

Hint: If you use a for loop here, you might create several variables identical to others. You can make another procedure to get rid of duplicates

$
\text{sex} + \left( \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} + \text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + \text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} \right)^2
$

```python
#variables
predictors_3 = sin_salario[['sex', 'exp1', 'exp2', 'exp3', 'exp4','hsg', 'scl', 'clg', 'ad', 'so', 'we', 'ne', 'occ2', 'ind2']]
#obtenemos dummies
predictors_3 = pd.get_dummies(predictors_3, columns=['occ2', 'ind2'], drop_first=True, dtype=int)

# Todas las columnas excepto 'sex'
columns_to_include = [col for col in predictors_3.columns if col != 'sex']

# suma de todas las variables seleccionadas
predictors_3['sum_vars'] = predictors_3[columns_to_include].sum(axis=1)

# Nueva variable:cuadrado de la suma de todas las variables
predictors_3['sum_vars_squared'] = predictors_3['sum_vars'] ** 2

display(predictors_3.head())
```

```python
# Convertir el DataFrame a un array NumPy
predictors3_array = predictors_3.to_numpy()

# Mostrar las columnas del DataFrame final
print("Columnas del DataFrame final:")
print(predictors_3.columns)

# Mostrar el contenido del array NumPy
print("Contenido del array NumPy:")
print(predictors3_array)

# Mostrar la forma del array NumPy
print("Forma del array NumPy:")
print(predictors3_array.shape)
```

9. Normalize all the variables, including $Y$. This means that for each variable $V$, create $\tilde{V}$ such that

$
\tilde{V} = \frac{V - \text{mean}(V)}{\text{sd}(V)}
$

```python
# Dataframe normalizada
normalized_data = data.copy()
# Normalizar todas las columnas numéricas
for column in normalized_data.select_dtypes(include=['float64', 'int64']).columns:
    mean = normalized_data[column].mean()
    std_dev = normalized_data[column].std()
    normalized_data[column] = (normalized_data[column] - mean) / std_dev

# Verificar el DataFrame normalizado
print(normalized_data.head())
```
