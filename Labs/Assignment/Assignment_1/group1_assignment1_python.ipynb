8. Make three new arrays for our predictors:

8.1. The basic model will have the columns sex hsg scl clg ad so we ne exp1 occ2 ind2. Make sure to convert occ2 and ind2 to dummies and to drop the first dummy value to prevent multicolinearity.

$
\text{sex} + \text{exp1} + \text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + \text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)}
$

```python
# Obtenermos las dummies
predictors = data[['sex', 'hsg', 'scl', 'clg', 'ad', 'so', 'we', 'ne', 'exp1', 'occ2', 'ind2']]
predictors = pd.get_dummies(predictors, columns=['occ2', 'ind2'],  drop_first=True, dtype=int)

#display(predictors.head())
# display(data.tail())
# display(predictors.info())
display(predictors.columns)
display(predictors.shape)

# Array
array_basic_model = predictors.values
display(array_basic_model.shape)
```

8.2. The flexible model will have the same columns, and will also include polynomials for experience (exp2 exp3 exp4), as well as the interactions between all experience variables and other variables except for sex. 

Hint: you can use a for loop to multiply the desired variables and create the interactions. Some packages might also have ways to easily specify the model variables with strings like R does by default

$
\text{sex} + \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} +
\text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + 
\text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} +
\left( \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} \right) \times 
\left( \text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + 
\text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} \right)
$

```python
# Variables
predictors_2= pd.concat([predictors, data[['exp2', 'exp3', 'exp4']]], axis=1)

desired_order = [
    'sex', 'exp1','exp2', 'exp3', 'exp4','hsg', 'scl', 'clg', 'ad', 'so', 'we', 'ne',  
    'occ2_2', 'occ2_3', 'occ2_4', 'occ2_5', 'occ2_6', 'occ2_7', 'occ2_8', 'occ2_9',
    'occ2_10', 'occ2_11', 'occ2_12', 'occ2_13', 'occ2_14', 'occ2_15', 'occ2_16',
    'occ2_17', 'occ2_18', 'occ2_19', 'occ2_20', 'occ2_21', 'occ2_22',
    'ind2_3', 'ind2_4', 'ind2_5', 'ind2_6', 'ind2_7', 'ind2_8', 'ind2_9', 'ind2_10',
    'ind2_11', 'ind2_12', 'ind2_13', 'ind2_14', 'ind2_15', 'ind2_16', 'ind2_17',
    'ind2_18', 'ind2_19', 'ind2_20', 'ind2_21', 'ind2_22'
]

# Order the columns
predictors_2 = predictors_2[desired_order]

# Array 1
array_flex1_model = predictors_2.values


# INTERACTIONS
experience_vars = ['exp1', 'exp2', 'exp3', 'exp4']
other_vars = [col for col in predictors.columns if col not in ['sex', 'exp1']] # exp1 is already on array_flex1_model

# List for interactions
interactions = []

# LOOP FOR INTERACTIONS
for exp_var in experience_vars:
    for other_var in other_vars:
        interaction_var = predictors_2[exp_var] * predictors_2[other_var]
        interactions.append(interaction_var.values.reshape(-1, 1))

# ARRAY
array_interactions = np.hstack(interactions)
array_flexible_model = np.hstack([array_flex1_model, array_interactions])

display(array_flexible_model.shape)
```

8.3. The extra-flexible model will include all two-way interactions between variables, except for sex. 

Hint: If you use a for loop here, you might create several variables identical to others. You can make another procedure to get rid of duplicates
    
$
\text{sex} + \left( \text{exp1} + \text{exp2} + \text{exp3} + \text{exp4} + \text{hsg} + \text{scl} + \text{clg} + \text{ad} + \text{so} + \text{we} + \text{ne} + \text{dummy(occ2)} + \text{dummy(ind2)} \right)^2
$

```python
# Variables
predictors_3 = predictors_2.copy()

# Array 1
array_extr_flex1_model = predictors_3.values

# Not including 'sex'
extra_flexible_vars = [col for col in predictors_3.columns if col != 'sex']

# List for interactions
interactions = []

# INTERACTIONS
for i, var1 in enumerate(extra_flexible_vars):
    for var2 in extra_flexible_vars[i+1:]:
        # Crear la variable de interacción
        interaction_var = predictors_3[var1] * predictors_3[var2]
        # Añadir la variable de interacción a la lista
        interactions.append(interaction_var.values.reshape(-1, 1))
        
# Array of interactions
interactions_array = np.hstack(interactions)

# Final array
array_extra_flexible_model = np.hstack([predictors_3[['sex']].values, predictors_3[extra_flexible_vars].values, interactions_array])
display(array_extra_flexible_model.shape)
```

R_{adj}^{2} = 1-\frac{n}{n-p} \frac{MSE}{TSS}
