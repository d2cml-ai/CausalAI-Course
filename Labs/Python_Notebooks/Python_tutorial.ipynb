# If condition

- Equals: a == b
- Not Equals: a != b
- Less than: a < b
- Less than or equal to: a <= b
- Greater than: a > b
- Greater than or equal to: a >= b

<img src="https://www.dropbox.com/s/8xwqk2ze1116qvp/if-statement.jpg?dl=1" alt="image info" />

```python
A = 20
B = 30

if A == B:
    print( 'Test expresion 1 is True' )
    
elif A > B:
    print( 'Test expresion 2 is True' )

elif A >= B:
    print( 'Test expresion 3 is True' )
    
elif A < B:
    print( 'Test expresion 4 is True' )
    
elif A <= B:
    print( 'Test expresion 5 is True' )
    
else:
    print( 'Test expresion 6 is True' )
```

# Linear Regression Model

```python
## loading the data
Penn = pd.read_csv("../data/penn_jae.dat" , sep='\s', engine='python')
```

```python
Penn.shape
```

```python
Penn = Penn[Penn.tg.isin([0,4])]

Penn
 #subset(Penn, tg==4 | tg==0)
```

```python
Penn = Penn.replace({"tg":4}, 1)  # replace 4 to 1 
Penn = Penn.rename(columns = {"tg": "T4"})
```

```python

import statsmodels.formula.api as smf
```

```python
# Formula regression
formula1 = "T4 ~ female" 

formula2 = "T4 ~ female + black"

# We include only the interaction

formula3 = "T4 ~ female + black + female:black"

# We include the two terms and their interaction

formula4 = "T4 ~ female*black"

# drop intercep

formula5 = "T4 ~ -1 + female + black + female:black"

# polinomial independent varibles

formula6 = "T4 ~ -1 + female + black + female:black + lusd + lusd^2 "

# interaction effects & categorical variables 

formula7 = "T4~(female+black+othrace+C(dep)+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd)**2"
```

```python
# Create category variable
Penn['dep'] = Penn['dep'].astype( 'category' )
```

```python
Penn['dep'].dtype
```

```python
mod = smf.ols(formula1, data=Penn).fit()
print(mod.summary())
```

```python
mod = smf.ols(formula3, data=Penn).fit()

print(mod.summary())
```

```python
mod = smf.ols(formula4, data=Penn).fit()
print(mod.summary())
```

```python
mod = smf.ols(formula6, data=Penn).fit()
print(mod.summary())
```

```python
mod = smf.ols(formula7, data=Penn).fit()
print(mod.summary())
```

```python
Penn = Penn[['female', 'black', 'lusd' , 'dep', 'T4']]

formula_modelB = "T4 ~  female + black + lusd + lusd^2+ C(dep)"
```

```python
mod = smf.ols(formula_modelB, data = Penn).fit()

print("Parameters: ", mod.params,'\n')

print("R2:", mod.rsquared,'\n')

print("Standard error:", mod.bse,'\n')

print("Predictived values:", mod.predict())
```

```python
new_obs = {'female': [0], 'black': [0], 'dep': [1], 'lusd': [1]}

new_obs = pd.DataFrame(new_obs)
new_obs
```

```python
y_hat  =  mod.predict(new_obs)
y_hat
```
