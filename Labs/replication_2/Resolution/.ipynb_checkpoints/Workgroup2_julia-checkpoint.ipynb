## Prediction Question

```julia
#basic model
basic  = @formula(lwage ~ (sex + exp1 + shs + hsg + mw + so + we + occ2+ ind2))
basic_results  = lm(basic, data)
```

```julia
beta_basic = GLM.coeftable(basic_results).cols[1][2]
beta_basic_se = GLM.coeftable(basic_results).cols[2][2]
println("Coefficient for basic model " , beta_basic, "standard error:", beta_basic_se)
```

```julia
# couples variables combinations 
combinations_upto(x, n) = Iterators.flatten(combinations(x, i) for i in 1:n)

# combinations without same couple
expand_exp(args, deg::ConstantTerm) =
    tuple(((&)(terms...) for terms in combinations_upto(args, deg.n))...)

StatsModels.apply_schema(t::FunctionTerm{typeof(^)}, sch::StatsModels.Schema, ctx::Type) =
    apply_schema.(expand_exp(t.args_parsed...), Ref(sch), ctx)
```

```julia
# Flexible model
    
flex_model = @formula(lwage ~ sex + (exp1+exp2+exp3+exp4+shs+hsg+occ2+ind2+mw+so+we)^2)
flex_model = apply_schema(flex_model, schema(flex_model, data))
```

```julia
flex_results = lm(flex_model, data)
```

```julia
beta_flexible = GLM.coeftable(flex_results).cols[1][2]
beta_flexible_se = GLM.coeftable(flex_results).cols[2][2]
```

```julia
println("Coefficient for flexible model " , beta_flexible, "standard error:", beta_flexible_se)
```

```julia
table = NamedArray(zeros(2, 5))

table[1,2] = GLM.coeftable(basic_results).cols[1][2]
table[1,3] = GLM.coeftable(basic_results).cols[2][2]
table[1,4] = GLM.coeftable(basic_results).cols[5][2]
table[1,5] = GLM.coeftable(basic_results).cols[6][2]

table[2,2] = GLM.coeftable(flex_results).cols[1][2]
table[2,3] = GLM.coeftable(flex_results).cols[2][2]
table[2,4] = GLM.coeftable(flex_results).cols[5][2]
table[2,5] = GLM.coeftable(flex_results).cols[6][2]

T = DataFrame(table, [ :"Model", :"Estimate", :"standar error", :"Lower bound", :"Upper bound"])  # table to dataframe 
T[!,:Model] = string.(T[!,:Model])  # string - first column 

T[1,1] = "OLS"
T[2,1] = "OLS Flexible"


header = (["Model", "Estimate", "standar error", "Lower bound", "Upper bound"])

pretty_table(T; backend = Val(:html), header = header, formatters=ft_round(4), alignment=:c)
```

```julia
xerror = T[!,2] .- T[!,4]

scatter( T[!,2], T[!,1], label = "", xerrors = xerror, 
        xtick = -1:0.01:1, linestyle = :dash, seriescolor=:red)

plot!(size=(750,400), title="Linear Model - CI")
```

### Assess the predictive performance


```julia
# R squared
r2_basic = r2(basic_results)
r2_flexible = r2(flex_results)

# R squared_ajusted
r2adj_basic = adjr2(basic_results)
r2adj_flexible = adjr2(flex_results)
```

```julia
y = data[!,:lwage]

#Prediction 

y_predict_basic = predict(basic_results)
y_predict_flex = predict(flex_results)

# MSE
mse_basic = mse(y_predict_basic, y)
mse_flexible = mse(y_predict_flex, y)

n = length(y)  # total observations
coef = GLM.coef(flex_results)

p1 = length(GLM.coef(basic_results))  # number of regresors
p2 = length(coef[coef .!= 0])  # number of regresors

# R MSE_ajusted
mseadj_basic = n/(n-p1)*mse(y_predict_basic, y)
mseadj_flexible = n/(n-p2)*mse(y_predict_flex, y)
```

```julia
coefnames(flex_model)
Y = select(data,:lwage)  # uptcome variable
col = coefnames(flex_model)[2]  # regresors control 
```

```julia
X = StatsModels.modelmatrix(flex_model,data)
X = DataFrame(X, col)
```

```julia
Y = select(data,:lwage) 

shrinkage_model = rlasso_arg( X, Y, nothing, true, true, true, false, false, 
                    nothing, 1.1, nothing, 5000, 15, 10^(-5), -Inf, true, Inf, true )
```

```julia
model_penalize = rlasso(shrinkage_model)
```

```julia
sum(model_penalize["residuals"].^2)
```

```julia
p3 = size(model_penalize["coefficients"])[1]
N = size(Y)[1]
```

```julia
scr = sum(model_penalize["residuals"].^2)
r2_lasso = 1- scr/model_penalize["rss"]
r2_adj_lassso = 1- ((N-1)/(N-p3))*scr/model_penalize["rss"]
```

```julia
MSE_lasso = mean(model_penalize["residuals"].^2)
MSE_adj_lasso = (N/(N-p3))*MSE_lasso
```

```julia
table = NamedArray(zeros(3, 6))

table[1,2] = p1
table[1,3] = r2_basic
table[1,4] = r2_flexible
table[1,5] = mse_basic
table[1,6] = mseadj_basic

table[2,2] = p2
table[2,3] = r2_flexible
table[2,4] = r2adj_flexible
table[2,5] = mse_flexible
table[2,6] = mseadj_flexible

table[2,2] = p3
table[3,3] = r2_lasso
table[3,4] = r2_adj_lassso
table[3,5] = MSE_lasso
table[3,6] = MSE_adj_lasso

T = DataFrame(table, [ :"Model", :"p", :"R2", :"R2_ajusted", :"MSE", :"MSE_ajusted"])  # table to dataframe 
T[!,:Model] = string.(T[!,:Model])  # string - first column 

T[1,1] = "Basic OLS"
T[2,1] = "Felxible OLS"
T[3,1] = "Felxible Lasso"

header = (["Model", "p", "R2", "R2_ajusted", "MSE", "MSE_ajusted"])

pretty_table(T; backend = Val(:html), header = header, formatters=ft_round(4), alignment=:c)
```
